```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
import torchvision
import torchvision.transforms as transforms
import numpy as np
import copy
import random
import matplotlib.pyplot as plt
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, 1, 1), nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, 3, 1, 1), nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(64*7*7, 128), nn.ReLU(),
            nn.Linear(128, 10)
        )
    def forward(self, x):
        return self.net(x)
```

```python
transform = transforms.Compose([transforms.ToTensor()])
```

```python
mnist_train = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
mnist_test = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)
train_data, valid_data = torch.utils.data.random_split(mnist_train, [50000, 10000])
```

```python
def partition_clients(trainset, num_clients=10):
    client_size = len(trainset) // num_clients
    client_indices = np.arange(len(trainset))
    np.random.shuffle(client_indices)
    partitions = [client_indices[i*client_size:(i+1)*client_size] for i in range(num_clients)]
    return partitions
```

```python
num_clients = 10
byzantine_frac = 0.3
byzantine_clients = random.sample(range(num_clients), int(num_clients*byzantine_frac))
client_partitions = partition_clients(train_data, num_clients=num_clients)
trusted_loader = data.DataLoader(valid_data, batch_size=200, shuffle=True)
trusted_data_iter = iter(trusted_loader)
trusted_images, trusted_labels = next(trusted_data_iter)
trusted_set = list(zip(trusted_images, trusted_labels))
```

```python
def train_local(model, client_data, device='cpu', lr=0.01, epochs=1):
    loader = data.DataLoader(client_data, batch_size=32, shuffle=True)
    model.train()
    optimizer = optim.SGD(model.parameters(), lr=lr)
    loss_fn = nn.CrossEntropyLoss()
    for _ in range(epochs):
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            loss = loss_fn(model(x), y)
            loss.backward()
            optimizer.step()
    return copy.deepcopy(model.state_dict())
```

```python
def malicious_update(model, client_data, device='cpu'):
    weights = train_local(model, client_data, device)
    for k in weights:
        weights[k] = torch.randn_like(weights[k])
    return weights
```

```python
def fedavg_aggregate(client_weights):
    agg = copy.deepcopy(client_weights[0])
    for k in agg:
        agg[k] = torch.mean(torch.stack([cw[k] for cw in client_weights]), dim=0)
    return agg
```

```python
def get_flat_weights(state_dict):
    return torch.cat([v.flatten() for v in state_dict.values()])

def set_flat_weights(model, flat_tensor):
    pointer = 0
    new_state = {}
    for k, v in model.state_dict().items():
        num = v.numel()
        new_state[k] = flat_tensor[pointer:pointer+num].reshape(v.shape).clone()
        pointer += num
    return new_state

def fltrust_aggregate(client_weights, server_model, trusted_set, device='cpu'):
    server_copy = copy.deepcopy(server_model)
    base_weights = get_flat_weights(server_copy.state_dict())
    train_local(server_copy, trusted_set, device, epochs=1)
    server_trust_update = get_flat_weights(server_copy.state_dict()) - base_weights

    aggregate = torch.zeros_like(server_trust_update)
    scores = []
    updates = []
    for cw in client_weights:
        flat_update = get_flat_weights(cw) - base_weights
        score = torch.cosine_similarity(flat_update, server_trust_update, dim=0)
        norm_update = server_trust_update.norm() * flat_update / (flat_update.norm() + 1e-9)
        updates.append(norm_update)
        scores.append(score)
    scores = torch.tensor(scores)
    scores = torch.relu(scores)
    scores = scores / (scores.sum() + 1e-12)
    for update, alpha in zip(updates, scores):
        aggregate += alpha * update
    new_weights = base_weights + aggregate
    return set_flat_weights(server_model, new_weights)
```

```python
def evaluate(model, dataloader, device='cpu'):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            out = model(x)
            pred = torch.argmax(out, 1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return correct / total
```

```python
def run_federated(aggregate_fn, method_name, num_rounds=10, local_epochs=1):
    server_model = SimpleCNN()
    train_accs, val_accs, test_accs = [], [], []
    for r in range(num_rounds):
        client_weights = []
        for c_idx, c_inds in enumerate(client_partitions):
            client_data = data.Subset(train_data, c_inds)
            model = copy.deepcopy(server_model)
            if c_idx in byzantine_clients:
                weights = malicious_update(model, client_data)
            else:
                weights = train_local(model, client_data, epochs=local_epochs)
            client_weights.append(weights)
        if method_name == "FLTrust":
            new_state_dict = aggregate_fn(client_weights, server_model, trusted_set)
        elif method_name == "FedAvg":
            new_state_dict = aggregate_fn(client_weights)
        server_model.load_state_dict(new_state_dict)
        train_loader = data.DataLoader(train_data, batch_size=128)
        val_loader = data.DataLoader(valid_data, batch_size=128)
        test_loader = data.DataLoader(mnist_test, batch_size=128)
        train_accs.append(evaluate(server_model, train_loader))
        val_accs.append(evaluate(server_model, val_loader))
        test_accs.append(evaluate(server_model, test_loader))
        print(f"[{method_name}] Round {r+1}/{num_rounds}: Val Acc = {val_accs[-1]:.4f}, Test Acc = {test_accs[-1]:.4f}")
    return train_accs, val_accs, test_accs
```

```python
fedavg_results = run_federated(fedavg_aggregate, "FedAvg")
```

```python
fltrust_results = run_federated(fltrust_aggregate, "FLTrust")
```

```python
plt.figure(figsize=(8, 4))
plt.plot(fedavg_results[2], label='FedAvg', linestyle='--')
plt.plot(fltrust_results[2], label='FLTrust', linestyle='-')
plt.xlabel('Communication Round')
plt.ylabel('Test Accuracy')
plt.legend()
plt.title('FLTrust vs FedAvg under Byzantine Attack')
plt.show()
```
